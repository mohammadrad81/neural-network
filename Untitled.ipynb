```python
from src.activations import LinearActivation
from src.layers.fullyconnected import FC
from src.losses.meansquarederror import MeanSquaredError
from src.model import Model
from src.optimizers.gradientdescent import GD
```

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("datasets/california_houses_price/california_housing_train.csv")
```

```python
data
```

```python
X_train, y_train = data.drop("median_house_value", axis=1), data[["median_house_value"]]
```

```python
X_train
```

```python
y_train
```

```python
X_train_arr = np.array(X_train)
```

```python
y_train_arr = np.array(y_train)
```

```python
y_train_arr = y_train_arr / np.max(y_train_arr)
```

```python
y_train_arr
```

```python
X_train_arr
```

```python
y_train_arr
```

```python
X_train_arr.shape
```

```python
fc_1 = FC(input_size=8, output_size=1, name='fc1')
```

```python
linear = LinearActivation()
```

```python
layers_list=[
    ('fc1', fc_1),
    ('linear', linear),
]
```

```python
model = Model(
    layers_list,
    criterion=MeanSquaredError(),
    optimizer=GD(layers_list=layers_list, learning_rate=0.01)
)
```

```python
model.train(X_train_arr, y_train_arr, epochs=10)
```
